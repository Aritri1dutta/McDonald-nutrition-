```python
##Analyze the distribution of calorie counts across menu items: This is done by creating a histogram chart.


plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Calorie Counts Across Menu Items')
plt.xlabel('Calories')
plt.ylabel('Frequency')

mean_calories = np.mean(df['Calories'])
median_calories = np.median(df['Calories'])
std_dev_calories = np.std(df['Calories'])


plt.text(150, 80, f'Mean Calories: {mean_calories:.2f}', fontsize=12, color='red')
plt.text(150, 75, f'Median Calories: {median_calories:.2f}', fontsize=12, color='blue')
plt.text(150, 70, f'Standard Deviation: {std_dev_calories:.2f}', fontsize=12, color='green')

plt.show()
```

# Create box plot for Carbohydrates content

```python
#box plot for distribution of carbohydratedcontent
plt.figure(figsize=(10,9))
sns.boxplot(x= 'Category', y='Carbohydrates',hue= 'Category', data=df, palette='rainbow')
plt.title('Box Plot of Carbohydrates counts Across Menu Categories')
plt.xlabel('Menu Categories')
plt.ylabel('Carbohydrates counts')
plt.xticks(rotation=45, ha='right')

plt.show()
```

Box plot showing carbohydrates content  in menue

```python
#average calories count
Calories=df.groupby(["Category"])["Calories"].mean()
Sorted_Calories=Calories.sort_values(ascending=False)
print(Sorted_Calories)
```

```python
plt.figure(figsize=(10,3))
sns.swarmplot(x='Category',y='Calories',data=df,palette='colorblind')
plt.title("Total mean according to Calories")
plt.xticks(rotation=45,ha='right')
plt.xlabel("Category")
plt.ylabel("Calories")
```
